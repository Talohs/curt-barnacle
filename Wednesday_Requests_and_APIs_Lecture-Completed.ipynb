```python
class WeatherAPI:
    base_url = 'http://api.weatherapi.com/v1'
    
    def __init__(self, api_key):
        self.api_key = api_key
        
    def _get(self, city, api_method, **params):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={city}"
        if params:
            for key, value in params.items():
                request_url += f"&{key}={value}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print('There was an error')
        
    def get_current_weather(self, city):
        weather_data = self._get(city, '/current.json')
        city_name = weather_data['location']['name']
        region_name = weather_data['location']['region']
        current_temp = weather_data['current']['temp_f']
        feels_like = weather_data['current']['feelslike_f']
        current_condition = weather_data['current']['condition']['text']
        city_weather = CityWeather(current_temp, feels_like, current_condition, city_name, region_name)
        return city_weather
            
        
class CityWeather:
    def __init__(self, current, feels_like, condition, city_name, region_name):
        self.current = current
        self.feels_like = feels_like
        self.condition = condition
        self.city = city_name
        self.region = region_name
        
    def __str__(self):
        degree_sign = u'\N{DEGREE SIGN}'
        return f"It is currently {self.condition} and {self.current}{degree_sign}F in {self.city}, {self.region}. \
It feels like {self.feels_like}{degree_sign}F"
        
    
client = WeatherAPI('1ac9901266fa4ab48f9185441220510')
my_city = client.get_current_weather('Chicago')
```

```python
class PokeAPI:
    base_url = "https://pokeapi.co/api/v2" 

    def _get(self, pokemon):
        request_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
        poke_response = requests.get(request_url)
        if poke_response.ok == True:
            return poke_response.json()
        else:
            print(f"There was a {poke_response.status_code} error")
    def get_pokedex(self):
        pokedex = 'https://pokeapi.co/api/v2/pokemon'
        poke_resp = requests.get(pokedex)
        if poke_resp.ok == True:
            poke_entries = poke_resp.json()
        else:
            print(f"There was a {pokedex.status_code} error")
        poke_entry = poke_entries['results']['name']
        return poke_entry

    def get_pokemon_info(self, pokemon):
        poke_stats = self._get(pokemon)
        pokemon = poke_stats['species']['name']
        poke_height = poke_stats['height']
        poke_weight = poke_stats['weight']
        pokemon_stats = PokemonHW(pokemon, poke_height, poke_weight)
        return pokemon_stats

class PokemonHW:
    def __init__(self, pokemon, height, weight):
        self.pokemon = pokemon
        self.height = height
        self.weight = weight

    def __str__(self):
        return f"{self.pokemon.title()} is {self.height} meters tall and weights {self.weight} kilograms"

# catch = PokeAPI()
# my_new_pokemon = catch.get_pokemon_info('pikachu')
# print(my_new_pokemon)

def catch_that_pokemon():
    catch = PokeAPI()
    pokedex = catch.get_pokedex()
    print(pokedex)
    pokemon = ''
    while True:
        pokemon = input("What pokemon did you spot? return to leave the grass.").lower()
        if pokemon in pokedex:
            my_new_pokemon = catch.get_pokemon_info(pokemon)
            print("You caught"+ my_new_pokemon)
            input("Do you want to catch another pokemon? return to leave the grass.").lower()
        else:
            pokemon = input("Please enter a pokemon? return to leave the grass").lower()
        if pokemon == 'return':
            break
        

catch_that_pokemon()
        
```
